```python
# ë°±ê·¸ë¼ìš´ë“œ ì‹¤í–‰
options = wb.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
```

```python
df_title = []
df_price = []
# í´ë” ê²½ë¡œê°€ì„œ \ -> / ë¡œ ë³€ê²½
# ë˜ë„ë¡ ì˜ì–´ë¡œ í• ê²ƒ option(ì†ì„±) = options(ìœ„ ë³€ìˆ˜)
driver = wb.Chrome("C:/Users/smhrd/í¬ë¡¤ë§/chromedriver.exe" , options = options)
time.sleep(0.5)
driver.get("http://corners.gmarket.co.kr/Bestsellers")
time.sleep(0.5)

for i in range(3) :
    img = driver.find_elements_by_css_selector("img.lazy")
    img[i].click()
    time.sleep(1)
    title = driver.find_element_by_css_selector(".itemtit")
    price = driver.find_element_by_css_selector(".price_real")
    df_title.append(title.text)
    df_price.append(price.text)
    driver.back()
    time.sleep(1)
```

```python
import pandas as pd
```

```python
# ë°ì´í„°í”„ë ˆì„ ì œì‘
df_number = []
for i in range(len(df_title)) :
    df_number.append(i+1)
gmarket = {'ìƒí’ˆì´ë¦„':df_title , 'ìƒí’ˆê°€ê²©':df_price , 'ìƒí’ˆë²ˆí˜¸': df_number}
df = pd.DataFrame(gmarket)
df.set_index('ìƒí’ˆë²ˆí˜¸' , inplace = True)
df
```
